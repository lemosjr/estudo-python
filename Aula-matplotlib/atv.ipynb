## Mais exemplos com pyplot

Continuando com a biblioteca Matplotlib, vamos praticar mais com funções como `plt.plot()`, `plt.bar()` e `plt.scatter()`. Lembre-se de usar rótulos, títulos e legendas para melhorar a clareza dos gráficos.

Funções úteis:

- `plt.plot()`: Para gráficos de linhas.
- `plt.bar()`: Para gráficos de barras.
- `plt.scatter()`: Para gráficos de dispersão.
- `plt.xlabel()`, `plt.ylabel()`, `plt.title()`: Para adicionar textos.
- `plt.show()`: Para exibir o gráfico.


## Crie um gráfico de linhas para mostrar a evolução da população de uma cidade ao longo dos anos

Use os dados de população anual de uma cidade hipotética de 2010 a 2020.


```python
import matplotlib.pyplot as plt

anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
populacao = [50000, 51000, 52000, 53500, 55000, 57000, 59000, 61000, 63000, 65000, 67000]

plt.plot(anos, populacao)
plt.xlabel('Anos')
plt.ylabel('população')
plt.title('Ano/pupulação')
plt.grid(True)
plt.show()
```

## Crie um gráfico de barras para comparar o número de livros lidos por mês

Dados de livros lidos mensalmente por uma pessoa ao longo de um semestre.


```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
livros = [3, 4, 2, 5, 6, 4]

plt.bar(meses, livros)
plt.xlabel('Mês')
plt.ylabel('Livros')
plt.title('Livros/Mês')
plt.show
```

## Crie um gráfico de dispersão para analisar a relação entre temperatura e vendas de sorvete

Dados de temperatura diária e vendas correspondentes.


```python
import matplotlib.pyplot as plt

temperaturas = [20, 22, 25, 28, 30, 32, 35, 37, 38, 40]
vendas_sorvete = [50, 55, 70, 85, 100, 120, 140, 150, 160, 180]

plt.scatter(temperaturas, vendas_sorvete)
plt.plot(temperaturas, vendas_sorvete)
plt.xlabel('Temperatura(c°)')
plt.ylabel('Vendas/Sorvete')
plt.show

```
