# Interface pyplot

## Introdução ao pyplot e suas funções

Matplotlib é uma biblioteca de plotagem em Python que produz figuras de qualidade em uma variedade de formatos impressos e ambientes interativos. `matplotlib.pyplot` é um conjunto de funções, e é geralmente importado sob o alias `plt`.

Cada função `pyplot` faz alguma alteração em uma figura, como criar uma figura, criar uma área de plotagem em uma figura, plotar algumas linhas em uma área de plotagem, decorar a plotagem com rótulos, etc.

Aqui estão algumas funções `pyplot` comumente usadas:

- `plt.figure()`: Cria uma nova figura.
- `plt.plot()`: Plota y versus x como linhas e/ou marcadores.
- `plt.scatter()`: Plota um gráfico de dispersão de y versus x.
- `plt.bar()`: Plota um gráfico de barras.
- `plt.xlabel()`: Define o rótulo do eixo x.
- `plt.ylabel()`: Define o rótulo do eixo y.
- `plt.title()`: Define um título para os eixos.
- `plt.grid()`: Configura as linhas da grade.
- `plt.legend()`: Coloca uma legenda no eixo.
- `plt.show()`: Exibe uma figura.


## Crie um gráfico de linhas para mostrar o crescimento das vendas de uma empresa ao longo do tempo

Suponha que temos dados de vendas mensais de uma empresa ao longo de um ano. Crie um gráfico de linhas para mostrar o crescimento das vendas:



```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
plt.plot(meses,vendas)
plt.xlabel('meses')
plt.ylabel('vendas')
plt.grid(True)
plt.show
```

## Crie um gráfico de barras para comparar as temperaturas médias de diferentes cidades

Vamos supor que temos dados de temperatura média de três cidades: São Paulo, Rio de Janeiro e Salvador. Crie um gráfico de barras para comparar essas temperaturas:


```python
import matplotlib.pyplot as plt

cidades = ['São Paulo', 'Rio de Janeiro', 'Salvador']
temperaturas = [20, 25, 27]

plt.bar(cidades,temperaturas)
plt.xlabel('cidades')
plt.ylabel('temperaturas(c°)')
plt.show
```

Suponha que temos dados do número de horas que alunos estudaram ao longo de uma semana para uma determinada prova e suas notas correspondentes. Crie um gráfico de dispersão para visualizar a correlação entre essas duas variáveis:
